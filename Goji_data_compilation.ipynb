```python
#Necessary packages:
!pip install --upgrade -q gspread

import gspread
import pandas as pd
from google.colab import auth
from google.auth import default
auth.authenticate_user()
creds, _ = default()
```

```python
Final = pd.DataFrame()
Final.index = ['Total', 'Carbon Emmisions', "Water Usage","Ethical Sourcing", "Labor Rights", "Transparency & Policy", "DEI"]
#Sheets file containing all of the links to individual company's information:
data_url = 'https://docs.google.com/spreadsheets/d/1oEaZVl3YF6JCxS63wsk10HO2dCVFs_CyfZ-cmlMFU0c/edit#gid=0'.replace("/edit#gid=", "/export?format=csv&gid=")
df2 = pd.read_csv(data_url, header = None)
```

```python
#Creating final table of data from all the companies:
for x in range(len(df2)): 
  x = df2.iloc[x,1]
  urlcur = x.replace("/edit#gid=", "/export?format=csv&gid=")
  dfcur = pd.read_csv(urlcur, header = None)
  total = dfcur.iloc[68,9]
  carbon = dfcur.iloc[84,9]
  water = dfcur.iloc[85,9]
  sourcing = sum(map(int,dfcur.iloc[[80,81],9]))
  labor = sum(map(int,dfcur.iloc[[71,72,73,76,79,92],9]))
  trans = sum(map(int,dfcur.iloc[[69,75,83,87,89,91,92],9]))
  dei = dfcur.iloc[90,9]
  Final[dfcur.iloc[2,2]] = [total, carbon, water, sourcing, labor, trans, dei]
```
